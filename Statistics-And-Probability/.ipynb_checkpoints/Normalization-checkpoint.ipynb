# Normal and Zed Distrobutions

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creat a list of data with a mean of 70 and a std of 10
a = np.random.normal(70, 20, 10000)
sns.distplot(a, hist=True, kde=True, bins=20)
```

```python
# Scale a to have a mean of 0
a_scaled = [(i-np.mean(a))/np.std(a) for i in a]
sns.distplot(a_scaled, hist=True, kde=True, bins=20)
```

## What is Zed distrobution?
### A ditrobution that is normal and has a mean of 0 and a std of 1

```python
# Function to calculate the normal distrobution of a given list fo values
def gaussian(x, mu, sigma):
    return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/2/(sigma**2))
```

```python
x = np.linspace(-5,5,100)
y = [gaussian(i, 0, 1) for i in x]

plt.plot(x,y)
plt.grid()
```
