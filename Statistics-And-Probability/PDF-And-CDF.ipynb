# Probability Density Function (PDF)

```python
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
```

```python
# Distrobution with kde set to false
sns.distplot(df['Age'].dropna(), hist=True, kde=False, bins=16)
```

```python
# Disity with kde set to true
sns.distplot(df['Age'].dropna(), hist=True, kde=True, bins=16)
```

## What is the probability of a passenger being younger than 40?

```python
ls_age = df['Age'].dropna().values

def percentage_lower(ls, lower):
    return sum(1 for value in ls if value < 40)/len(ls)

percentage_lower(ls_age, 40)
```

# Cumulative Density Function (CDF)

```python
import numpy as np
import matplotlib.pyplot as plt

dice = np.random.randint(low=1.0, high=7.0, size=1000)
```

```python
def cumlative_density(ls):
    amounts = []
    ls.sort()
    for value in range(0, int(max(ls))):
        amount_sum = sum(1 for i in ls if i <= value)
        amounts.append(amount_sum/len(ls))
    return amounts
```

```python
# CDF of dice
CDF_dice = cumlative_density(dice)
plt.plot(range(0, 6), CDF_dice)
plt.grid()
```

```python
# CDF of titanic age
CDF_list = cumlative_density(ls_age)

plt.plot(range(0, 80), CDF_list)
plt.grid()
```
